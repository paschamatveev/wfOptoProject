```python
# Time array
plt.figure(figsize=(5,2))
t = np.linspace(0, 1, 1000)  # Time from 0 to 1 second

# Step current
I = np.ones_like(t)  # Create an array of ones
I[t < 0.1] = 0  # Set the current to 0 for t < 0.5 seconds
I[t == 0.1] = 1.5  # Set the current to 1.5 at t = 0.5 seconds
I[t>.9]=0

# Plot
plt.plot(t, I)
plt.xlabel('Time (s)')
plt.ylabel('Current (mV)')
plt.title('Step Current')
plt.ylim([0,1.5])
plt.show()
```
